# Model Selection
 **Selecting the best model involves balancing interpretability, performance, and data characteristics.**
1. **Data Complexity:** Non-linear models (e.g., Decision Trees, Random Forest) handle complex relationships better, while Linear Regression works for simpler data.
2. **Overfitting Control:** Regularized models (Ridge, Lasso) and ensemble methods (Random Forest, Gradient Boosting) help mitigate overfitting.
3. **Interpretability vs. Accuracy:** Linear models are easy to interpret, while more complex models (e.g., Gradient Boosting) provide higher accuracy but less transparency.

### Available Machine Learning Model Options for Salary Prediction:

1. **Linear Regression:** Predicts salary based on a linear relationship between features and the target variable.
2. **Multiple Linear Regression:** Considers multiple features (e.g., experience, education) to improve salary prediction accuracy.
3. **Ridge Regression:** Regularizes the linear model to prevent overfitting, useful for handling multicollinearity.
4. **Lasso Regression:** Performs both regularization and feature selection by shrinking coefficients of irrelevant features to zero.
5. **Decision Trees:** Captures non-linear relationships by splitting data into branches based on feature values.
6. **Random Forest:** An ensemble of decision trees that reduces overfitting and captures complex patterns.
7. **Gradient Boosting:** Sequentially builds decision trees, improving model accuracy but requires cacharacteristics.

### Best Performing Models and Possible Reasons:

1. **Lasso Regression:**
   - **Reason:** Lasso Regression not only regularizes the model to prevent overfitting but also performs automatic feature selection, keeping only the most relevant features. This makes it ideal for datasets with many predictors, ensuring the model is both accurate and interpretable by reducing noise from irrelevant variables.

2. **Random Forest:**
   - **Reason:** Random Forest's ensemble approach aggregates multiple decision trees, reducing overfitting and capturing complex, non-linear relationships between features like experience, education, and salary. Its ability to handle both categorical and numerical data effectively makes it well-suited for this prediction task.

3. **Gradient Boosting:**
   - **Reason:** Gradient Boosting sequentially builds decision trees by focusing on correcting previous errors, resulting in a highly accurate model. It works particularly well with complex datasets and can capture intricate patterns, making it powerful for salary prediction when feature interactions are critical.

These models perform well due to their ability to manage complex data patterns, control overfitting, and maintain a balance between accuracy and generalization.ralization.

### Data Preprocessing

```python
# copying the df to keep the actual data safe from any change
df2 = df.copy()
df2.head()
```

* Categorical data cannot typically be directly handled by machine learning algorithms, as most algorithms are primarily designed to operate with numerical data only. Therefore, before categorical features can be used as inputs to machine learning algorithms, they must be encoded as numerical values.

```python
# for assigning the value according to tier converting the tier in list 
Tier1 = clg['Tier 1'].to_list()
Tier2 = clg['Tier 2'].to_list()
Tier3 = clg['Tier 3'].to_list()
```

```python
# Assign tier values to colleges
# - If 'Tier1' then 3
# - If 'Tier2' then 2
# - If 'Tier3' then 1
for item in df2['College']:
  if item in Tier1:
    df2['College'] = df2['College'].replace(item, 3) 
  elif item in Tier2:
    df2['College'] = df2['College'].replace(item, 2) 
  elif item in Tier3:
    df2['College'] = df2['College'].replace(item, 1) 
```

```python
df2.head()
```

```python
# assigning value to cities
# for metro 1
# for non metro 0
for item in df2['City']:
  if item in cty['Metrio City'].to_list():
    df2['City'] = df2['City'].replace(item, 1)
  elif item in cty['non-metro cities'].to_list():
    df2['City'] = df2['City'].replace(item, 0)
```

```python
df2.head()
```

```python
# creating dummy variable for 'Role' column 
df2 = pd.get_dummies(df2, columns=['Role'],drop_first=True,dtype=int)
df2.head()
```

# Apply the Machine Learning Models

### Lasso Regression 

```python
# Spliting the X and Y for regression
x= df2.drop('CTC', axis=1)
y= df2['CTC']
```

##### Without Scaling

```python
# importing all libraries required for lasso regression without Scaling
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
# spliting the data for train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) #here data is divided into 80%:train_data and 20%:test_data ratio
# checking the shape of the train and test data
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# applying lasso regression
lasso = Lasso()
lasso.fit(x_train, y_train)
```

```python
# checking mean_squared_error,r2_score,mean_absolute_error
print('r2_score:', r2_score(y_test, lasso.predict(x_test)))
print('mean_squared_error:', mean_squared_error(y_test, lasso.predict(x_test)))
print('mean_absolute_error:', mean_absolute_error(y_test, lasso.predict(x_test)))
```

```python
# with setting the alpha value
lso = Lasso(alpha=0.5)
lso.fit(x_train,y_train)
```

```python
# checking mean_squared_error,r2_score,mean_absolute_error
print('r2_score:', r2_score(y_test, lso.predict(x_test)))
print('mean_squared_error:', mean_squared_error(y_test, lso.predict(x_test)))
print('mean_absolute_error:', mean_absolute_error(y_test, lso.predict(x_test)))
```

##### With Scaling

```python
# applying lasso regression after scaling the data
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(x_train)
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
lsos =Lasso()
lsos.fit(x_train_scaled,y_train)
```

```python
# checking mean_squared_error,r2_score,mean_absolute_error
print('r2_score:', r2_score(y_test, lsos.predict(x_test_scaled)))
print('mean_squared_error:', mean_squared_error(y_test, lsos.predict(x_test_scaled)))
print('mean_absolute_error:', mean_absolute_error(y_test, lsos.predict(x_test_scaled)))
```

```python
#with setting alpha
lsosw = Lasso(alpha=0.5)
lsosw.fit(x_train_scaled,y_train)
```

```python
# checking mean_squared_error,r2_score,mean_absolute_error
print('r2_score:', r2_score(y_test, lsosw.predict(x_test_scaled)))
print('mean_squared_error:', mean_squared_error(y_test, lsosw.predict(x_test_scaled)))
print('mean_absolute_error:', mean_absolute_error(y_test, lsosw.predict(x_test_scaled)))
```

### Random Forest

* As we know Scaling features is generally not required for tree-based models like Random Forest and Gradient Boosting, so I am focusing on tuning the parameters instead.

##### without tunned parameters  

```python
# importing the RandomForestRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
# creating the RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train, y_train)
```

```python
# Evaluate the model by making predictions on the test data
print('r2_score:', r2_score(y_test, rfr.predict(x_test)))
print('mean_squared_error:', mean_squared_error(y_test, rfr.predict(x_test)))
print('mean_absolute_error:', mean_absolute_error(y_test, rfr.predict(x_test)))
```

##### with tunned parameters

```python
# creating the RandomForestRegressor with tunned parameter
rfrt = RandomForestRegressor(n_estimators=10000,  # Increase the number of trees
    max_depth=None, # Allow trees to grow to their full depth
    random_state=42,n_jobs=-1)
rfrt.fit(x_train, y_train)
```

```python
# Evaluate the model by making predictions on the test data
print('r2_score:', r2_score(y_test, rfrt.predict(x_test)))
print('mean_squared_error:', mean_squared_error(y_test, rfrt.predict(x_test)))
print('mean_absolute_error:', mean_absolute_error(y_test, rfrt.predict(x_test)))
```
